```python
def job_priority(title):
    title = title.lower()
    if 'ceo' in title or 'chief' in title or 'president' in title:
        return 10
    elif 'vp' in title or 'vice' in title or 'director' in title or 'head' in title:
        return 9
    elif 'senior' in title or 'principal' in title or 'lead' in title:
        return 8
    elif 'manager' in title or 'supervisor' in title:
        return 7
    elif any(word in title for word in ['engineer', 'scientist', 'analyst', 'developer', 'consultant']):
        return 6
    elif any(word in title for word in ['specialist', 'designer', 'coordinator', 'officer', 'recruiter']):
        return 5
    elif any(word in title for word in ['associate', 'assistant', 'representative']):
        return 4
    elif 'junior' in title or 'intern' in title:
        return 3
    elif any(word in title for word in ['clerk', 'trainee', 'apprentice']):
        return 2
    elif any(word in title for word in ['driver', 'receptionist']):
        return 1
    else:
        return 5
data_01['Job_Title_priority_encoded'] = data_01['Job Title'].apply(job_priority)
print(data_01[['Job Title', 'Job_Title_priority_encoded']].sample(20))
```

# Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold = KFold(n_splits=9, shuffle=True, random_state=42)
cv_scores = cross_val_score(linear_model, x_train, y_train, cv=kfold)
mse_scores = -cross_val_score(linear_model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
mae_scores = -cross_val_score(linear_model, x_train, y_train, cv=kfold, scoring='neg_mean_absolute_error')
print("Cross-Validation Scores:", cv_scores)
print("Mean Squared Error (CV):", mse_scores.mean())
print("Mean Absolute Error (CV):", mae_scores.mean())
```

```python
y_pred_linear= linear_model.predict(x_test)
print("r2 score(Linear) is=",r2_score(y_test,y_pred_linear))
print("Mean Squared Error (Linear):", mean_squared_error(y_test, y_pred_linear))
print("Mean Absolute Error (Linear):", mean_absolute_error(y_test, y_pred_linear))
```

# Polynomial Regression (Degree=2,3,4,5) And find the best model

```python
from sklearn.model_selection import KFold, cross_val_score
# ==============================
# K-FOLD SETUP
# ==============================
kfold = KFold(n_splits=9, shuffle=True, random_state=42)

# ==============================
# DEGREE RANGE 25
# ==============================
degrees = [2, 3, 4, 5]

results = []  # to store all CV results

for degree in degrees:
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    x_train_poly = poly.fit_transform(x_train)

    # Define and train model
    poly_reg_model = LinearRegression()
    poly_reg_model.fit(x_train_poly, y_train)

    # Cross-validation scores
    r2_scores = cross_val_score(poly_reg_model, x_train_poly, y_train, cv=kfold, scoring='r2')
    mse_scores = -cross_val_score(poly_reg_model, x_train_poly, y_train, cv=kfold, scoring='neg_mean_squared_error')
    mae_scores = -cross_val_score(poly_reg_model, x_train_poly, y_train, cv=kfold, scoring='neg_mean_absolute_error')

    # Store results
    results.append({
        'Degree': degree,
        'R2_Mean': r2_scores.mean(),
        'MSE_Mean': mse_scores.mean(),
        'MAE_Mean': mae_scores.mean()
    })

    # Print for each degree
    print(f"\n===================== DEGREE {degree} =====================")
    print(f"R (mean): {r2_scores.mean():.4f}")
    print(f"MSE (mean): {mse_scores.mean():.4f}")
    print(f"MAE (mean): {mae_scores.mean():.4f}")
    print("===========================================================")

# ==============================
# SUMMARY TABLE
# ==============================
results_df = pd.DataFrame(results)
print("\n Summary of Polynomial Regression (9-Fold CV):")
print(results_df)

# Identify the best performing degree (highest R)
best_degree = results_df.loc[results_df['R2_Mean'].idxmax(), 'Degree']
print(f"\n Best Degree based on CV R: {best_degree}")
```

# Lasso Regression (alpha=0.0001, 0.001, 0.01, 0.1, 10, 100) Find the best model

```python

# ==============================
# K-FOLD SETUP
# ==============================
kfold = KFold(n_splits=9, shuffle=True, random_state=42)

# ==============================
# ALPHA VALUES TO TEST
# ==============================
alphas = [0.0001, 0.001, 0.01, 0.1, 10, 100]

results = []  # to store all CV results

# ==============================
# LOOP THROUGH EACH ALPHA
# ==============================
for alpha in alphas:
    # Define and train model
    lasso_model = Lasso(alpha=alpha, random_state=42, max_iter=10000)

    # Cross-validation scores
    r2_scores = cross_val_score(lasso_model, x_train, y_train, cv=kfold, scoring='r2')
    mse_scores = -cross_val_score(lasso_model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
    mae_scores = -cross_val_score(lasso_model, x_train, y_train, cv=kfold, scoring='neg_mean_absolute_error')

    # Store results
    results.append({
        'Alpha': alpha,
        'R2_Mean': r2_scores.mean(),
        'MSE_Mean': mse_scores.mean(),
        'MAE_Mean': mae_scores.mean()
    })

    # Print for each alpha
    print(f"\n===================== ALPHA {alpha} =====================")
    print(f"R (mean): {r2_scores.mean():.4f}")
    print(f"MSE (mean): {mse_scores.mean():.4f}")
    print(f"MAE (mean): {mae_scores.mean():.4f}")
    print("===========================================================")

# ==============================
# SUMMARY TABLE
# ==============================
results_df = pd.DataFrame(results)
print("\n Summary of Lasso Regression (9-Fold CV):")
print(results_df)

# Identify the best performing alpha (highest R)
best_alpha = results_df.loc[results_df['R2_Mean'].idxmax(), 'Alpha']
print(f"\n Best Alpha based on CV R: {best_alpha}")
```

# Best Random Forest Model finding

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
# ==============================
# K-FOLD SETUP
# ==============================
kfold = KFold(n_splits=9, shuffle=True, random_state=42)

# ==============================
# HYPERPARAMETERS TO TEST
# ==============================
# (You can modify or expand these lists as needed)
n_estimators_list = [100]      # Number of trees
max_depth_list = [5, None]       # Tree depth
min_samples_split_list = [2, 5]      # Minimum samples to split
min_samples_leaf_list = [1, 4]        # Minimum samples per leaf

# ==============================
# RESULTS STORAGE
# ==============================
results = []

# ==============================
# LOOP THROUGH EACH COMBINATION
# ==============================
for n_est in n_estimators_list:
    for depth in max_depth_list:
        for split in min_samples_split_list:
            for leaf in min_samples_leaf_list:

                # Define model
                rf_model = RandomForestRegressor(
                    n_estimators=n_est,
                    max_depth=depth,
                    min_samples_split=split,
                    min_samples_leaf=leaf,
                    random_state=42,
                    n_jobs=-1
                )

                # Cross-validation scores
                r2_scores = cross_val_score(rf_model, x_train, y_train, cv=kfold, scoring='r2')
                mse_scores = -cross_val_score(rf_model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
                mae_scores = -cross_val_score(rf_model, x_train, y_train, cv=kfold, scoring='neg_mean_absolute_error')

                # Store results
                results.append({
                    'n_estimators': n_est,
                    'max_depth': depth,
                    'min_samples_split': split,
                    'min_samples_leaf': leaf,
                    'R2_Mean': r2_scores.mean(),
                    'MSE_Mean': mse_scores.mean(),
                    'MAE_Mean': mae_scores.mean()
                })

                # Print progress
                print(f"\n===== n_estimators={n_est}, max_depth={depth}, "
                      f"min_split={split}, min_leaf={leaf} =====")
                print(f"R (mean): {r2_scores.mean():.4f}")
                print(f"MSE (mean): {mse_scores.mean():.4f}")
                print(f"MAE (mean): {mae_scores.mean():.4f}")
                print("===========================================================")

# ==============================
# SUMMARY TABLE
# ==============================
results_df = pd.DataFrame(results)
print("\n Summary of Random Forest (9-Fold CV):")
print(results_df)

# Identify the best combination (highest R)
best_params = results_df.loc[results_df['R2_Mean'].idxmax()]
print("\n Best Hyperparameters based on CV R:")
print(best_params)
```

# Gradient Boosting

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
# ==============================
# K-FOLD SETUP
# ==============================
kfold = KFold(n_splits=9, shuffle=True, random_state=42)

# ==============================
# HYPERPARAMETERS TO TEST
# ==============================
learning_rate_list = [0.1]     # Shrinks contribution of each tree
n_estimators_list = [200]         # Number of boosting stages (trees)
max_depth_list = [5, 7]                  # Depth of individual estimators
min_samples_split_list = [2,10]         # Minimum samples to split
min_samples_leaf_list = [1, 4]           # Minimum samples at leaf

# ==============================
# RESULTS STORAGE
# ==============================
results = []

# ==============================
# LOOP THROUGH EACH COMBINATION
# ==============================
for lr in learning_rate_list:
    for n_est in n_estimators_list:
        for depth in max_depth_list:
            for split in min_samples_split_list:
                for leaf in min_samples_leaf_list:

                    # Define model
                    gbr_model = GradientBoostingRegressor(
                        learning_rate=lr,
                        n_estimators=n_est,
                        max_depth=depth,
                        min_samples_split=split,
                        min_samples_leaf=leaf,
                        random_state=42
                    )

                    # Cross-validation scores
                    r2_scores = cross_val_score(gbr_model, x_train, y_train, cv=kfold, scoring='r2')
                    mse_scores = -cross_val_score(gbr_model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
                    mae_scores = -cross_val_score(gbr_model, x_train, y_train, cv=kfold, scoring='neg_mean_absolute_error')

                    # Store results
                    results.append({
                        'learning_rate': lr,
                        'n_estimators': n_est,
                        'max_depth': depth,
                        'min_samples_split': split,
                        'min_samples_leaf': leaf,
                        'R2_Mean': r2_scores.mean(),
                        'MSE_Mean': mse_scores.mean(),
                        'MAE_Mean': mae_scores.mean()
                    })

                    # Print progress
                    print(f"\n===== learning_rate={lr}, n_estimators={n_est}, "
                          f"max_depth={depth}, min_split={split}, min_leaf={leaf} =====")
                    print(f"R (mean): {r2_scores.mean():.4f}")
                    print(f"MSE (mean): {mse_scores.mean():.4f}")
                    print(f"MAE (mean): {mae_scores.mean():.4f}")
                    print("===========================================================")

# ==============================
# SUMMARY TABLE
# ==============================
results_df = pd.DataFrame(results)
print("\n Summary of Gradient Boosting (9-Fold CV):")
print(results_df)

# Identify the best combination (highest R)
best_params = results_df.loc[results_df['R2_Mean'].idxmax()]
print("\n Best Hyperparameters based on CV R:")
print(best_params)
```
